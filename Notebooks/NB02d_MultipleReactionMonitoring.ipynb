# Multiple Reaction Monitoring (MRM)

Multiple Reaction Monitoring (MRM) is a targeted mass spectrometric acquisition strategy that is primarily applied for the sensitive and selective quantification of predefined analytes and is widely used in targeted proteomics, metabolomics, and pharmaceutical analysis.

In contrast to data-dependent acquisition approaches, which aim to identify as many peptides as possible in a sample, MRM focuses exclusively on a defined set of precursor ions and their corresponding fragment ions. This targeted principle enables highly reproducible and quantitative measurements, even in complex biological matrices.

In the context of bottom-up proteomics, MRM is typically applied to proteotypic peptides that uniquely represent a protein of interest. By monitoring multiple transitions per peptide, the reliability of identification and quantification is further increased. Thus, MRM represents a robust and well-established strategy for targeted protein quantification in complex biological samples. (Neil R, et al, 2009)



For a MRM approach the triple quadrupole is often the mass analyzer of choice it consists of three quadrupoles arranged in series (Q1–Q2–Q3). 

- In the first quadrupole (Q1), a specific precursor ion with a defined mass-to-charge ratio (m/z) is selected from all ions entering the mass spectrometer. 
- This precursor ion is then transferred into the second quadrupole (Q2), which functions as a collision cell. 
- There, the ion undergoes collision-induced dissociation (CID) through energetic collisions with an inert gas, resulting in characteristic fragment ions. 
- The third quadrupole (Q3) subsequently filters for one specific fragment ion derived from the selected precursor. Only this predefined fragment ion is transmitted to the detector. 




(Gabriela V, et al, 2012)

The combination of one selected precursor ion in Q1 and one selected fragment ion in Q3 is referred to as a transition. In Multiple Reaction Monitoring, several such transitions are monitored sequentially within one analytical run. Each transition is specific for a particular peptide or analyte and therefore enables highly selective detection.
The strength of MRM lies in its double mass filtering (in Q1 and Q3), which significantly reduces chemical noise and increases specificity. As a consequence, low-abundance peptides can be quantified with high sensitivity and precision. For quantitative applications, stable isotope-labeled internal standards are often used to allow accurate relative or absolute quantification.
